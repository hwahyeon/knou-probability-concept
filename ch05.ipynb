```r
library(distrEx)
```

### 1. 초기하분포

```r
X = Hyper(2,2,2)
plot(X)
```

```r
# 초기하분포의 확률질량함수값
d(X)(0:2)

# 기댓값과 분산
cat(" E(X)      = ", E(X), "\n")
cat(" var(X)    = ", var(X), "\n")
```

### 2. 이항분포

```r
# 이항분포를 이용한 확률계산
X=Binom(4,0.8)
plot(X)
```

```r
d(X)(2)
# 기댓값과 분산
cat(" E(X)      = ", E(X), "\n")
cat(" var(X)    = ", var(X), "\n")
```

### 3. 포아송분포

```r
X=Pois(1)
plot(X)
```

```r
# 포아송 분포를 이용한 확률계산
1-p(X)(2)

# 이항 분포를 이용한 확률계산
X1=Binom(10000,0.0001)
1-p(X1)(2)
```

```r
# 기댓값과 분산
cat(" E(X)      = ", E(X), "\n")
cat(" var(X)    = ", var(X), "\n")
```
