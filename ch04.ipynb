```r
#install.packages("distrEx")
library(distrEx)
```

```r
# 주사위 던지기의 분포
X =  DiscreteDistribution ( supp = c(1:6),  prob = rep(1/6,6) )
plot(X)
```

```r
# 기댓값과 분산(X=주사위던지기 눈금)
cat(" E(X)      = ", E(X), "\n")
cat(" var(X)    = ", var(X), "\n")
cat(" sd(X)     = ", sd(X), "\n")
cat(" E(2*X+5)  = ", E(2*X+5), "\n")
cat(" var(2*X+5)= ", var(2*X+5), "\n")
cat(" sd(2*X+5) = ", sd(2*X+5))
```
