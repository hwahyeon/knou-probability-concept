### 1. 연속형 균등분포

```r
X = Unif(1,5)
plot(X)
```

```r
# 확률 계산 : P(X>2)
1-p(X)(2)
# 확률 계산 : P(2<=X<4)
p(X)(4) - p(X)(2)
```

```r
# 기댓값과 분산
cat(" E(X)      = ", E(X), "\n")
cat(" var(X)    = ", var(X), "\n")
```

### 2. 지수분포

```r
X = Exp(3)
plot(X)
```

```r
# 확률계산 : P(X<1)
 p(X)(1)
```

```r
# 기댓값과 분산
cat(" E(X)      = ", E(X), "\n")
cat(" var(X)    = ", var(X), "\n")
```

### 3. 정규분포

```r
X = Norm(15,5)
plot(X)
```

```r
# 확률계산 : P(X<25)
p(X)(25)
```

```r
# 기댓값과 분산
cat(" E(X)      = ", E(X), "\n")
cat(" var(X)    = ", var(X), "\n")
```
